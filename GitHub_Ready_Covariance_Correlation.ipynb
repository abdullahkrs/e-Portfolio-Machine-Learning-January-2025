# ðŸ“ˆ Covariance and Pearson Correlation Analysis using cars.csv

This notebook analyzes the statistical relationship between **Engine Volume** and **CO2 Emissions** using the `cars.csv` dataset. It demonstrates how covariance and Pearson's correlation can be used to evaluate the strength and direction of a linear relationship between two numerical variables.

## ðŸ“Œ Description
- **Dataset**: Uploaded manually via Colab. Columns include `Volume`, `Weight`, `CO2`, etc.
- **Objective**: Measure how changes in engine volume affect CO2 output.
- **Metrics Used**:
  - Covariance
  - Pearson Correlation Coefficient
- **Visualization**: Scatterplot with regression line to aid visual understanding of correlation strength.

## ðŸ§ª Learning Outcomes
- Understand how changes in feature variables (like Volume) influence target variables (like CO2).
- Learn how correlation metrics are calculated and interpreted.
- Gain hands-on experience working with real-world automotive data.

---


```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import cov
from scipy.stats import pearsonr
from google.colab import files
import io

# Upload CSV file
uploaded = files.upload()
df = pd.read_csv(io.BytesIO(uploaded[list(uploaded.keys())[0]]))

# Display info
print("Dataset Head:")
print(df.head())

# Select variables
data1 = df['Volume']
data2 = df['CO2']

# Calculate covariance matrix
covariance = cov(data1, data2)
print("\nCovariance matrix:\n", covariance)

# Calculate Pearson's correlation
corr, _ = pearsonr(data1, data2)
print(f"\nPearson correlation coefficient between Volume and CO2: {corr:.4f}")

# Plot
sns.regplot(x=data1, y=data2)
plt.xlabel("Engine Volume (cc)")
plt.ylabel("CO2 Emissions (g/km)")
plt.title("Correlation between Volume and CO2")
plt.grid(True)
plt.show()
```

# üîÅ Polynomial Regression using cars.csv

This notebook demonstrates **Polynomial Regression** using the `cars.csv` dataset to model the non-linear relationship between a car‚Äôs **weight** and its **CO2 emissions**.

## üìå Description
- **Feature (x)**: `Weight` (kg)
- **Target (y)**: `CO2` (g/km)
- **Method**: `numpy.polyfit` with degree 2 (quadratic fit)

## üìà Objective
To explore whether a non-linear curve fits the data better than a simple straight-line regression.

## üß™ Learning Outcomes
- Understand the concept and application of polynomial regression
- Visualize complex relationships in data
- Compare curve fitting with linear regression

---


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
import io

# Upload CSV file
uploaded = files.upload()
df = pd.read_csv(io.BytesIO(uploaded[list(uploaded.keys())[0]]))

# Use Weight and CO2 for polynomial regression
x = df['Weight']
y = df['CO2']

# Fit a 2nd-degree polynomial model
mymodel = np.poly1d(np.polyfit(x, y, 2))

# Create line for plotting
myline = np.linspace(x.min(), x.max(), 100)

# Plot original points and polynomial regression curve
plt.scatter(x, y, label='Actual Data')
plt.plot(myline, mymodel(myline), color='red', label='Polynomial Regression (Degree 2)')
plt.xlabel("Car Weight (kg)")
plt.ylabel("CO2 Emissions (g/km)")
plt.title("Polynomial Regression: Weight vs CO2")
plt.grid(True)
plt.legend()
plt.show()
```

```python

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

#NumPy has a method that lets us make a polynomial model
mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))

#specify how the line will display, we start at position 1, and end at position 22
myline = numpy.linspace(1, 22, 100)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()
```

It is important to know how well the relationship between the values of the x- and y-axis is, if there are no relationship the polynomial regression can not be used to predict anything.

The relationship is measured with a value called the r-squared.

The r-squared value ranges from 0 to 1, where 0 means no relationship, and 1 means 100% related

```python
from sklearn.metrics import r2_score

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))

print(r2_score(y, mymodel(x)))
```

# ðŸ“Š Gradient Descent: Learning Rate & Iteration Comparison

This notebook builds on the `gradient_descent_cost_function.ipynb` tutorial by comparing:

- **Different iteration values** with a fixed learning rate
- **Different learning rates** with a fixed number of iterations

These visualizations help explain how gradient descent behaves under different hyperparameter settings.

---


```python

import numpy as np
import matplotlib.pyplot as plt

# Target minimum
target_theta = 2

# Cost function: J(theta) = (theta - target)^2
def cost(theta):
    return (theta - target_theta)**2

# Gradient descent for cost minimization
def gradient_descent(start_theta, learning_rate, iterations):
    theta = start_theta
    cost_history = []
    for _ in range(iterations):
        grad = 2 * (theta - target_theta)
        theta -= learning_rate * grad
        cost_history.append(cost(theta))
    return cost_history
```


## ðŸŒ€ Cost vs Iteration Comparison (Fixed Learning Rate)

This plot compares how the number of iterations affects convergence when using a constant learning rate of `0.01`.


```python

learning_rate = 0.01
theta = 0
iteration_options = [10, 50, 100, 300]
costs_by_iterations = {}

for iters in iteration_options:
    costs_by_iterations[iters] = gradient_descent(theta, learning_rate, iters)

plt.figure(figsize=(10, 6))
for iters, costs in costs_by_iterations.items():
    plt.plot(range(len(costs)), costs, label=f"{iters} Iterations")
plt.xlabel("Iteration")
plt.ylabel("Cost")
plt.title("Cost Reduction for Different Iteration Values (Learning Rate = 0.01)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```


## ðŸš€ Cost vs Learning Rate Comparison (Fixed Iteration Count)

This plot compares how different learning rates impact convergence behavior when using 100 fixed iterations.


```python

iteration_count = 100
start_theta = 0
learning_rates = [0.001, 0.01, 0.05, 0.1, 0.5]
costs_by_lr = {}

for lr in learning_rates:
    theta = start_theta
    cost_history = []
    for _ in range(iteration_count):
        grad = 2 * (theta - target_theta)
        theta -= lr * grad
        cost_history.append(cost(theta))
    costs_by_lr[lr] = cost_history

plt.figure(figsize=(10, 6))
for lr, costs in costs_by_lr.items():
    plt.plot(range(len(costs)), costs, label=f"LR = {lr}")
plt.xlabel("Iteration")
plt.ylabel("Cost")
plt.title("Cost Reduction with Varying Learning Rates (Iterations = 100)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```


## âœ… Summary

- A **higher number of iterations** improves convergenceâ€”up to a point.
- **Learning rate** must be carefully chosen:
  - Too small â†’ slow learning
  - Too large â†’ unstable or divergent

These hyperparameters greatly influence the **speed and stability** of gradient descent optimization.
